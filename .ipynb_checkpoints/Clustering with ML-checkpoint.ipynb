# Credit card clustering with Machine Learning

Le clustering des cartes de crédit est une technique d'analyse de données utilisée pour regrouper les titulaires de cartes de crédit en segments homogènes, basés sur divers attributs comportementaux associés à leur utilisation de la carte de crédit.

Une segmentation efficace peut aider les institutions financières à personnaliser leurs offres et leurs stratégies de marketing pour chaque segment (cluster), améliorant ainsi la satisfaction de la clientèle.

## 1. Qu'est-ce que le clustering ? 

Le clustering est une méthode d’apprentissage non supervisé. Il consiste à rassembler des données similaires ou proches 
en « clusters » ou en groupes distincts sans l’utilisation d’étiquettes préalables.

Il s’oppose ainsi à l’apprentissage supervisé qui se base sur des données étiquetées pour effectuer des prédictions. 

Dans notre analyse, nous utiliserons les K-means (l’un des algorithmes de clustering les plus populaires) pour regrouper
des individus dans des segments en fonctions leurs caractéristiques.

<b> Analyse en Composantes Principales (ACP) </b>

L'objectif de l'ACP est de réduire la dimensionnalité des données tout en conservant autant que possible la variance totale.

Nous utilisons l'ACP pour transformer les variables originales en un ensemble de nouvelles variables non corrélées 
(composantes principales) qui expliquent la majeure partie de la variabilité.

Nous conservons les composantes expliquant 90% de la variance.

La segmentation des individus se fera sur la base des composantes principales.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Standardisation des données
scaler = StandardScaler()
data_scaled = scaler.fit_transform(card_)
```

```python
# Application de l'ACP
pca = PCA(n_components=0.9)  # On garde les composantes principales expliquant 90% de la variance
data_pca = pca.fit_transform(data_scaled)

# Affichage de la variance expliquée par chaque composante principale
print("Variance expliquée par chaque composante principale :")
print(pca.explained_variance_ratio_)

# Visualisation de la variance expliquée cumulée
plt.figure()
plt.plot(range(1, len(pca.explained_variance_ratio_)+1), pca.explained_variance_ratio_.cumsum(), marker='o', linestyle='--')
plt.xlabel('Nombre de composantes principales')
plt.ylabel('Variance expliquée cumulée')
plt.title('Variance expliquée cumulée par les composantes principales')
plt.grid()
plt.show()
```

```python
# Extraction des composantes principales (vecteurs propres)
components = pca.components_

# Création d'un DataFrame pour les coefficients des composantes principales
components_df = pd.DataFrame(components, columns=card_.columns, index=[f'PC{i+1}' for i in range(components.shape[0])])
```

```python
print(components_df)
```

```python
# Récupération des noms de colonnes des données originales
columns_names = card_.columns  

# Boucle sur chaque composante principale
for i, component in enumerate(pca.components_):
    print(f"Composante principale {i + 1}:")
    # Trie les indices des variables par ordre de l'importance (valeur absolue des coefficients)
    sorted_indices = np.argsort(np.abs(component))[::-1]
    # Affichage des 5 variables contribuant le plus aux composantes principales
    for j in sorted_indices[:5]: 
        print(f"{columns_names[j]}: {component[j]}")
    print("--------------------")
```

Nous cherchons maintenant à détécter les variables les plus récurrentes sur l'ensemble des composantes afin de déterminer ce qui pourrait expliquer la particularité de chaque segment une fois le regroupement efféctué.

```python
from collections import Counter

# Récupération des noms de colonnes des données originales
columns_names = card_.columns  # Assurez-vous que 'data' est votre DataFrame ou tableau de données d'origine

# Initialisation d'un compteur pour les variables les plus importantes
important_vars_counter = Counter()

# Nombre de variables les plus importantes à considérer pour chaque composante
top_n = 5  # Par exemple, les 5 variables les plus importantes

# Boucle sur chaque composante principale
for i, component in enumerate(pca.components_):
    print(f"Composante principale {i + 1}:")
    # Trie les indices des variables par ordre de l'importance (valeur absolue des coefficients)
    sorted_indices = np.argsort(np.abs(component))[::-1]
    # Compte les variables les plus importantes pour cette composante
    for j in sorted_indices[:top_n]:
        important_vars_counter[columns_names[j]] += 1
        print(f"{columns_names[j]}: {component[j]}")
    print("--------------------")

# Affichage des variables les plus récurrentes
print("Variables les plus récurrentes parmi les plus importantes:")
for var, count in important_vars_counter.most_common():
    print(f"{var}: {count} fois")

```

```python
# Visualisation des résultats 
import matplotlib.pyplot as plt

# Tri des variables par récurrence
sorted_vars = [var for var, count in important_vars_counter.most_common()]

# Comptage des récurrences
counts = [count for var, count in important_vars_counter.most_common()]

plt.figure(figsize=(10, 6))
plt.barh(sorted_vars, counts, color='skyblue')
plt.xlabel('Nombre de récurrences parmi les composantes principales')
plt.title('Variables les plus récurrentes parmi les composantes principales')
plt.gca().invert_yaxis()  # Inversion de l'axe y pour que les variables les plus récurrentes soient en haut
plt.show()
```

## 4. Les K-means

Le K-means est un algorithme de clustering couramment utilisé pour regrouper des données non étiquetées en clusters distincts, 
en fonction de leurs similarités.

Les étapes principales du K-means se présentent comme suit:

* <b>Initialisation des centroïdes</b> : Le processus commence par choisir aléatoirement K centroïdes, où K est le nombre de clusters que vous souhaitez former. Les centroïdes sont des points dans l'espace des données qui serviront de centres initiaux pour chaque cluster.

-  <b> Affectation des points aux clusters </b> : Pour chaque point de données, le K-means calcule la distance par rapport à chaque centroïde et affecte le point au cluster dont le centroïde est le plus proche. La distance la plus couramment utilisée est la distance euclidienne, bien que d'autres mesures puissent être utilisées en fonction du domaine d'application.


*  <b> Mise à jour des centroïdes  </b> : Une fois que tous les points ont été affectés à des clusters, les centroïdes de chaque cluster sont mis à jour en calculant la moyenne de tous les points appartenant à ce cluster. Cela déplace le centroïde vers le centre géométrique du cluster nouvellement formé.


-  <b>Répétition des étapes 2 et 3  </b> : Les étapes d'affectation et de mise à jour sont répétées de manière itérative jusqu'à ce qu'un critère d'arrêt soit atteint. Le critère d'arrêt peut être un nombre fixe d'itérations, une convergence des centroïdes (c'est-à-dire que les centroïdes ne se déplacent plus de manière significative) ou une autre condition définie par l'utilisateur.


*  <b>Convergence et attribution finale </b> : Une fois que le processus converge, c'est-à-dire que les centroïdes ne changent plus de position, chaque point de données est assigné définitivement au cluster dont le centroïde est le plus proche.

### 4.1. Le choix du nombre de clusters

L'une des étapes les plus importantes de l'algorithme des kmeans est la détermination du nombre de cluster.

Pour choisir le nombre optimal de clusters, on lance souvent le k-means plusieurs fois, avec différentes valeurs de K.
Pour chacune d'entre elles, on calcule l’inertie intraclasse (within-cluster sum of squares, ou WCSS).
Plus il y a de clusters, plus le nombre d'individus par cluster diminue, et plus les clusters sont resserrés. Donc, bien entendu, l’inertie intraclasse diminue forcément quand K augmente.

Pour déterminer le nombre de clusters à étudier, on s'intéresse au graphique qui trace l'inertie intraclasse en fonction du nombre de clusters.
On peut déterminer le nombre idéal de clusters en appliquant **la méthode du coude**.


Dans le cadre de notre étude, nous avons choisi k=10 clusters. Pour chaque cluster, on entraine un k-means spécifique et on stocke son inertie.
On retient 

```python
from sklearn.cluster import KMeans

# Calcul du WCSS pour différents nombres de clusters
wcss = []
K_range = range(1, 11)  # Test de 1 à 10 clusters

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_pca)
    wcss.append(kmeans.inertia_) 
    
# Tracer la courbe du WCSS
plt.figure(figsize=(10, 4))
plt.plot(K_range, wcss, marker='o', linestyle='--')
plt.xlabel('Nombre de clusters (K)')
plt.ylabel('WCSS')
plt.title('La méthode du coude')
plt.grid(True)
plt.show()
```

On remarque qu'au delà de 4 clusters, la diminution de la WCSS commence à ralentir.

Nous retenons donc 4 groupes.

```python
kmeans = KMeans(n_clusters=4, random_state=42)
cluster_labels = kmeans.fit_predict(data_pca)
```

```python
# Ajout des labels de clusters au DataFrame original
card_['Cluster'] = cluster_labels

# Visualisation des clusters dans le plan formé par les deux premières composantes principales.
plt.figure(figsize=(8, 6))
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=cluster_labels, cmap='viridis')
plt.xlabel('Composante principale 1')
plt.ylabel('Composante principale 2')
plt.title('Clusters obtenus par K-means')
plt.colorbar(label='Cluster')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import plotly.express as px

# Visualisation des clusters en 3D avec Plotly
fig = px.scatter_3d(
    x=data_pca[:, 0], 
    y=data_pca[:, 1], 
    z=data_pca[:, 2], 
    color=cluster_labels.astype(str),  # Convertir les labels en chaîne de caractères pour l'affichage
    title='Clusters obtenus par K-means en 3D',
    labels={'x': 'Composante principale 1', 'y': 'Composante principale 2', 'z': 'Composante principale 3'},
    opacity=0.8
)

fig.update_layout(scene=dict(
    xaxis_title='Composante principale 1',
    yaxis_title='Composante principale 2',
    zaxis_title='Composante principale 3'
))

# Afficher la figure
fig.show()
```
